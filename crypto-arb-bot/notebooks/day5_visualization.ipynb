```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the processed CSV from Day 3
df = pd.read_csv('../data/ticker_analysis.csv')

# Convert timestamp column to datetime and set as index
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.set_index('timestamp', inplace=True)

# Display first few rows
df.head()
```

```python
print("Columns:", df.columns.tolist())
print("\nBasic stats:")
print(df.describe())
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['bid'], label='Bid', color='green')
plt.plot(df.index, df['ask'], label='Ask', color='red')
plt.plot(df.index, df['mid'], label='Mid', color='blue', linestyle='--')

plt.title('Bid–Ask–Mid Price Relationship')
plt.xlabel('Time (UTC)')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(df.index, df['spread'], color='purple', linewidth=2)
plt.title('Absolute Spread (Ask - Bid)')
plt.xlabel('Time (UTC)')
plt.ylabel('Spread (USD)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(df.index, df['rel_spread_%'], color='orange', linewidth=2)
plt.title('Relative Spread (%)')
plt.xlabel('Time (UTC)')
plt.ylabel('Spread (%)')
plt.grid(True)
plt.show()
```

```python
max_spread_row = df.loc[df['spread'].idxmax()]

plt.figure(figsize=(10, 4))
plt.plot(df.index, df['spread'], color='purple', linewidth=2)
plt.scatter(max_spread_row.name, max_spread_row['spread'], color='red', s=60, label='Max Spread')
plt.title('Spread (USD) with Max Spread Highlighted')
plt.xlabel('Time (UTC)')
plt.ylabel('Spread (USD)')
plt.legend()
plt.grid(True)
plt.show()

print(f"Max spread: {max_spread_row['spread']:.4f} USD at {max_spread_row.name}")
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['mid'], df['spread'], alpha=0.6, color='teal')
plt.title('Relationship between Mid Price and Spread')
plt.xlabel('Mid Price (USD)')
plt.ylabel('Spread (USD)')
plt.grid(True)
plt.show()
```
